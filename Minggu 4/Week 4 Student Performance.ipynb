# Visualisasi
Skor rata-rata dan plot menunjukkan bahwa siswa-siswa ini cenderung sedikit lebih baik dalam membaca dan menulis dibandingkan matematika

```python
df[["math score","reading score","writing score"]].describe()
```

```python
df.agg({'math score': ['min','mean','max', 'median', 'skew'],
             'writing score': ['min', 'max', 'median', 'mean','skew']})
```

```python
#Average math score for male and female comparison
df[["gender","math score"]].groupby("gender").mean()
```

# Visualisasi

Berdasarkan data, siswa laki-laki tampaknya lebih baik dalam matematika dibandingkan siswa perempuan

```python
df[["test preparation course","math score","reading score","writing score"]].groupby("test preparation course").mean()
```

# Visualiasi
Orang yang telah menyelesaikan kursus persiapan ujian tampaknya memiliki nilai yang lebih baik dibandingkan orang yang tidak menyelesaikan kursus.

```python
a = df[["race/ethnicity","math score","reading score","writing score"]].groupby("race/ethnicity").mean()
a
```

```python
a.plot()
```

# Visualisasi

Grup E tampaknya berkinerja lebih baik dibandingkan semua kelompok ras/etnis lainnya dalam semua kriteria kinerja

```python
b = df[["parental level of education","math score","reading score","writing score"]].groupby("parental level of education").mean()
b
```

```python
b.plot(figsize=(20,16))
```

# Visualisasi
Tingkat pendidikan orang tua tampaknya mempunyai pengaruh terhadap prestasi siswa. Misalnya saja siswa yang orang tuanya bergelar magister menunjukkan kinerja yang lebih baik dibandingkan siswa lainnya.

```python
c = df[["lunch","math score","reading score","writing score"]].groupby("lunch").mean()
c
```

```python
c.plot()
```
